# Setup

```python
import os
import numpy as np
import pandas as pd
from scipy import stats

AIRBNB_PATH = os.path.join("data", "airbnb")

def get_data_from_csv_file(file_name, aribnb_path=AIRBNB_PATH):
    path = os.path.join(aribnb_path, file_name)
    return pd.read_csv(path)
```

# Test T

### kmeans_cluster_based_on_numerical_and_categorical_variables_0

```python
airbnb_0 = get_data_from_csv_file("kmeans_cluster_based_on_numerical_and_categorical_variables_0.csv")
airbnb_0.head()
```

```python
airbnb_0["cluster"].value_counts()
```

```python
airbnb_0.groupby("cluster").mean()
```

```python
a = airbnb_0[airbnb_0["cluster"] == 0]["review_scores_rating"].values
b = airbnb_0[airbnb_0["cluster"] == 1]["review_scores_rating"].values

stats.ttest_ind(a, b)
```

```python
a = airbnb_0[airbnb_0["cluster"] == 0]["review_scores_cleanliness"].values
b = airbnb_0[airbnb_0["cluster"] == 1]["review_scores_cleanliness"].values

stats.ttest_ind(a, b)
```

```python
a = airbnb_0[airbnb_0["cluster"] == 0]["review_scores_value"].values
b = airbnb_0[airbnb_0["cluster"] == 1]["review_scores_value"].values

stats.ttest_ind(a, b)
```

```python
a = airbnb_0[airbnb_0["cluster"] == 0]["price"].values
b = airbnb_0[airbnb_0["cluster"] == 1]["price"].values

stats.ttest_ind(a, b)
```
